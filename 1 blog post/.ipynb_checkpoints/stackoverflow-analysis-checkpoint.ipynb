# Airline Delay Analysis

```python
data = pd.read_csv('airlines.csv')
data.head()
```

```python
data.columns
```

```python
data.isnull().sum()
```

## Q1: Which airport has the most number of total flights while with the highest delay？

```python
data['total_delay'] = data['Statistics.# of Delays.Carrier'] + data['Statistics.# of Delays.Late Aircraft'] + data['Statistics.# of Delays.National Aviation System'] + data['Statistics.# of Delays.Weather']
data['city'] = data['Airport.Name'].apply(lambda x: x.split(':')[0])
```

```python
q1 = data.groupby(['Time.Year','city'],as_index=False).agg('mean')[['Time.Year','city','Statistics.Flights.Total','total_delay']].sort_values(['Statistics.Flights.Total','total_delay'], ascending=False)
q1.head(30)
```

```python
plt.scatter(q1[q1['city']=='Atlanta, GA']['Time.Year'], q1[q1['city']=='Atlanta, GA']['total_delay'])
```

```python
plt.scatter(q1[q1['city']=='Seattle, WA']['Time.Year'], q1[q1['city']=='Seattle, WA']['total_delay'])
```

## Q2: Is weather the most main reason for delay?

```python
(data['Statistics.# of Delays.Weather'] / data['total_delay'] < 0.2).sum() / data.shape[0]
```
